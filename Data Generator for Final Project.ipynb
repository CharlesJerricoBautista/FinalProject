```python
import pandas as pd
from faker import Faker
import random
```

```python
def create_customer(num_customers):
    fake = Faker()
    customer_list = []
    for customer_id in range(1, num_customers + 1):
        customer = {
            'customer_id': customer_id,
            'first_name': fake.first_name(),
            'last_name': fake.last_name(),
            'billing_address': fake.address(),
            'shipping_address': fake.address(),
            'contact_number': fake.random_number(digits=10),
            'order_id' : fake.random_number(digits=3)}
        customer_list.append(customer)
    return pd.DataFrame(customer_list)
```

```python
create_customer(1000)
```

```python
customer_data = create_customer(1000)
```

```python
customer_data.to_csv('customer.csv')
```

```python
def create_order(num_order):
    fake = Faker()
    order_list = []
    for order_id in range(1, num_order + 1):
        order = {
            'order_id': fake.random_number(digits=3),
            'quantity_ordered': fake.random_number(digits=1),
            'order_date': fake.date_between(start_date='-1y', end_date='today'),
            'order_status': random.choice(['Processing', 'Shipped', 'Delivered']),
            'product_id': fake.random_number(digits=3),
            'total_order_amount': round(random.uniform(1, 1000), 2),
            'customer_id' : fake.random_number(digits=3)}
        order_list.append(order)
    return pd.DataFrame(order_list)
```

```python
create_order(1000)
```

```python
order_data = create_order(1000)
```

```python
order_data.to_csv('order.csv')
```

```python
def create_product(num_product):
    fake = Faker()
    product_list = []
    for product_id in range(1, num_product + 1):
        product = {
            'product_id': fake.random_number(digits=3),
            'product_name': fake.word(),
            'product_price': round(random.uniform(1, 1000), 2),
            'product_category': random.choice(['Drinks', 'Toys', 'Snacks']),
            'product_description': fake.sentence()}
        product_list.append(product)
    return pd.DataFrame(product_list)
```

```python
create_product(1000)
```

```python
product_data = create_order(1000)
```

```python
product_data.to_csv('product.csv')
```
